```python
import whisper

model = whisper.load_model("base")
```

```python
result = model.transcribe("Sports.wav")

print(result["text"])
```

```python
# cleaning the processed text

text = result["text"]

text = text.lower().strip()

print(text)
```

```python
# tokenization for future intent nlp

tokens = text.split()

print(tokens)
```
